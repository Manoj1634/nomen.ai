```python
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

model_id = "mistralai/Mistral-7B-Instruct-v0.3"

# ✅ Fix 1: Add bitsandbytes import
# Required for load_in_4bit=True
import bitsandbytes as bnb

# ✅ Fix 2: Specify torch_dtype explicitly to avoid precision mismatch errors
from torch import bfloat16  # or float16 if your GPU supports it better

# Load quantized model
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto",
    load_in_4bit=True,
    torch_dtype=bfloat16,
    trust_remote_code=True
)

# Load tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_id)

# Create generation pipeline
pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    device_map="auto",
    max_new_tokens=150,
    do_sample=True,
    temperature=0.7,
    top_k=50,
    top_p=0.95
)

# Test prompt
prompt = "Suggest three creative domain names for a vegan food delivery service:"
response = pipe(prompt)
print(response[0]["generated_text"])
```
